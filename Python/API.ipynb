# API => Application Programming Interface

```python
import requests
import time
import pandas as pd
```

```python
url = f"http://makeup-api.herokuapp.com/api/v1/products.json?product_type=lipstick"
resp = requests.get(url)
result = resp.json()
```

```python
type(result)
```

```python
df = pd.DataFrame(result)
df.head(3)
```

```python
for i in range (4,5):
    url_2 = f"http://makeup-api.herokuapp.com/api/v1/products.json?rating_greater_than={i}"
    resp_2 = requests.get(url_2)
    result_2 = resp_2.json()
    df_2 = pd.DataFrame(result_2)
    time.sleep(1)
```

```python
df_rating = df_2[['id', 'name','brand','category','price','rating']].sort_values(['rating'], ascending=False)
```

```python
df_rating.head(10)
```

```python
df_rating.tail(10)
```

```python
df_rating.dropna()
```

```python
df_high_rating_brand = df_rating[['id', 'brand']].groupby(['brand']).count()\
                       .rename(columns = {'id' : 'No_of_product_high-rating'})\
                       .sort_values(['No_of_product_high-rating'],ascending=False)
```
